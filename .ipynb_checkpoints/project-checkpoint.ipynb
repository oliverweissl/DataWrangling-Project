```python
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
start = "2023-01-01"
end = "2023-01-08"
```

```python
def percent_diff(col1,col2):
    return ((col2 - col1) / col1) * 100
```

```python
path = "Stocks.xlsx"
df = pd.read_excel(path)
del df['Unnamed: 1']
```

```python
df.head(5)
```

```python
# Only Tickers that are on NMS and GER
ny_ger = df.loc[df["NMS"].notnull() & df["GER"].notnull(), ["NMS", "GER"]]
ny_ger_tups = list(ny_ger.itertuples(index=False, name=None))

# download each pair of Tickers and plot them
for tup in ny_ger_tups:
    stocks = ' '.join([tup[0], tup[1]])
    data = yf.download(stocks, start, end, interval='1m')
    data['Open'].plot(figsize=(16,12))

# # get a list of all the Tickers to download
# stocks = ' '.join(' '.join([tup[0], tup[1]]) for tup in ny_ger_tups)
# data = yf.download(stocks, start, end, interval='1m')
```

```python
# Only Tickers that are on NYQ and HKG
ny_hkg = df.loc[df["NYQ"].notnull() & df["HKG"].notnull(), ["NYQ", "HKG"]]

ny_hkg_tups = list(ny_hkg.itertuples(index=False, name=None))

# download each pair of Tickers and plot them
for tup in ny_hkg_tups:
    stocks = ' '.join([tup[0], tup[1]])
    data = yf.download(stocks, start, end, interval='1m')
    data['Open'].plot(figsize=(16,12))
```
